```python
def BubbleSort(a):
    n = len(a)
    for i in range(0, n - 1):
        for j in range(i + 1, n):
            if (a[i] > a[j]):
                temp = a[i]
                a[i] = a[j]
                a[j] = temp
    return a
```

```python
a = [1, 3, -1, 6, 7, 0, 3]
print(*BubbleSort(a))
```

```python
n = 2000

import random
l = []
for i in range(n):
    l.append(random.randint(0, 100))
```

```python
import time

timers = []

for lens in range(1, n, 10):
    m = l[:lens]
    start = time.time()
    BubbleSort(m)
    end = time.time()
    timers.append(end - start)
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, n, 10), timers)
plt.plot(range(1, n, 10), [0.00000005 * i * i for i in range(1, n, 10)])
```

```python
def BubbleSort_elementary(a):
    n = len(a)
    count = 2
    count += 2
    flag = True
    for i in range(0, n - 1):
        count += 1
        for j in range(i + 1, n):
            count += 4
            if (a[i] > a[j]):
                flag = False 
                temp = a[i]
                a[i] = a[j]
                a[j] = temp
                count += 7
        if (flag):
            break
    return count
```

```python
#  

def SelectionSort(a):
    n = len(a)
    for i in range(0, n - 1):
        min_ind = i
        for j in range(i + 1, n):
            if (a[j] < a[min_ind]):
                min_ind = j
        a[i], a[min_ind] = a[min_ind], a[i]
    return a
```

```python
print(*SelectionSort([0, 9, 4, 3, 1]))
print(*SelectionSort([1, 2, 3, 4]))
print(*SelectionSort([9, 8, 7, 6, 5]))
print(*SelectionSort([0, 9, 0, 1, 2, 3]))
```
